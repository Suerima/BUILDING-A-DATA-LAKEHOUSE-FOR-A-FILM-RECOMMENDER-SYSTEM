```
info_df = spark.read.table('silver.movie').select('id','imdbId', 'title').sort('id')
info_df.show()
```

```
from delta.tables import *
from pyspark.sql.functions import col, substring

# Define the silver location
info_goldLocation = "/mnt/datalake/gold/dim_info"

# Table name
info_goldTable = "gold.Dim_Info"

info_df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save(info_goldLocation)
```

```
spark.sql(f"CREATE EXTERNAL TABLE IF NOT EXISTS {info_goldTable} USING delta LOCATION '{info_goldLocation}'")
```

```
# Maintenance for Delta table
info_DataDelta = DeltaTable.forName(spark, info_goldTable)

if info_DataDelta.history(30).filter("operation = 'VACUUM START'").count() == 0:
	info_DataDelta.optimize()
	info_DataDelta.vacuum()
```
